```python
import numpy
import math
import scipy.misc
from scipy.misc.pilutil import Image
import scipy.fftpack as fftim
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
%pylab inline

import sys
```

## Aqui voy a abrir la imagen y la voy a convertir a escala de grises. Posteriormente voy a guardar la imagen como un array bidimensional de python, y le voy a aplicar la transformada de fourier.

```python
filename = '20_popc_cho009-1.tif'
def procesar_imagen(filename):
    imagen = Image.open(filename).convert('L') #http://effbot.org/imagingbook/image.htm
    
    array_imagen = numpy.asarray(imagen) # Convert the input to an array. - http://docs.scipy.org/doc/numpy/reference/generated/numpy.asarray.html
    
    array_imagen_fft = (fft.fft2(array_imagen))  # Transformada de Fourier al array_imagen
    
    return array_imagen_fft
```

## Aqui muestro un plot de la Transformada de Fourier para la imagen '20_popc_cho009-1.tif', para mostrar porque razon hay que reorganizar las frecuencias

```python
F_Transform = procesar_imagen('20_popc_cho009-1.tif')
r,c = shape(F_Transform)


x = linspace(0,c-1,c)
y = linspace(0,r-1,r)

from mpl_toolkits.mplot3d import Axes3D

X,Y = meshgrid(x,y)




# `ax` is a 3D-aware axis instance because of the projection='3d' keyword argument to add_subplot
fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.set_zlim(0, 0.1*(10**7))
p = ax.plot_surface(X, Y, abs(F_Transform), rstride=4, cstride=4, linewidth=0)
plt.title('Transformada de Fourier rapida de la Imagen 20_popc_cho009-1.tif', fontsize=20 )
plt.show()


#fig.colorbar(surf, shrink=0.5, aspect=5)

```

## Como se puede observar en la imagen anterior, las frecuencias altas quedan concentradas en el centro X y Y del plot, mientras que en las esquinas encontramos las frecuencias mas pequenas.

## Por las razones anteriores aqui voy a voltear los datos obtenidos a traves de la funcion anterior. La funcion que utilice automaticamente me manda las frecuencias mas bajas a las esquinas del array, y yo quiero que las frecuencias mas bajas esten en en centro para poder cortarlas con un circulo.

```python
ff_order = F_Transform.copy()

size_filas = r # Gives the number of rows in the 2d Array
size_columnas = c# Gives the number of columns in the 2d Array

ff_order[ size_filas/2:size_filas , size_columnas/2:size_columnas ] = F_Transform[0:size_filas/2 , 0:size_columnas/2]
ff_order[ 0:size_filas/2 , 0:size_columnas/2] = F_Transform[size_filas/2:size_filas , size_columnas/2:size_columnas ]
ff_order[ 0:size_filas/2 , size_columnas/2:size_columnas ] = F_Transform[size_filas/2:size_filas , 0:size_columnas/2]
ff_order[size_filas/2:size_filas , 0:size_columnas/2] = F_Transform[0:size_filas/2 , size_columnas/2:size_columnas ]

fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.set_zlim(0, 0.1*(10**7))
p = ax.plot_surface(X, Y, abs(ff_order), rstride=4, cstride=4, linewidth=0)
plt.title('Transformada de Fourier rapida REORGANIZADA de la Imagen 20_popc_cho009-1.tif', fontsize=20 )
plt.show()
```

```python
def shift_fourier(filename):
    array_imagen_fft2 = procesar_imagen(filename) # Llamar a la funcion anterior para obtener un array
    
    shift_array = fft.fftshift(array_imagen_fft2) # Shift del array para organizar las frecuencias. http://docs.scipy.org/doc/numpy/reference/generated/numpy.fft.fftshift.html
    
    return shift_array
```

```python
def plot_fourier(filename):
    F_Transform = shift_fourier(filename)
    r,c = shape(F_Transform)


    x = linspace(0,c-1,c)
    y = linspace(0,r-1,r)

    

    X,Y = meshgrid(x,y)






    # `ax` is a 3D-aware axis instance because of the projection='3d' keyword argument to add_subplot
    fig = plt.figure(figsize=(14,6))

    ax = fig.add_subplot(1, 1, 1, projection='3d')

    ax.set_zlim(0, 0.1*(10**7))
    p = ax.plot_surface(X, Y, abs(F_Transform), rstride=4, cstride=4, linewidth=0)

    plt.show()
```

## Aqui voy a hacer la funcion filtro.

```python
def funcion_filtro_suave(filename):
    shift_array = shift_fourier(filename)
    #shift_array = procesar_imagen(filename)
    r = shift_array.shape[0]
    c = shift_array.shape[1]
    
    #r,c = shape(shift_array)
    
    M = numpy.ones((r,c))
    
    centro_eje_x = c/2
    centro_eje_y = r/2
    
    limite_frecuencias = 10
    
    t = 2*limite_frecuencias
    
    for i in range(1,r):
        for j in range(1,c):
            radio_al_centro = numpy.sqrt( (centro_eje_x - j)**2 + (centro_eje_y - i)**2 )
            
            if (radio_al_centro >= limite_frecuencias):
                M[i][j] = math.exp(-(radio_al_centro**2)/(t**2) )
    
    
    M = scipy.misc.toimage(M)
    
    array_filtrado = shift_array * M
    
    array_filtrado_inversa = fftim.ifft2(array_filtrado)
    
    abs_array_filtrado = abs(array_filtrado_inversa)
    
    imagen = scipy.misc.toimage(abs_array_filtrado)
    
    
    #Reference http://docs.opencv.org/modules/highgui/doc/reading_and_writing_images_and_video.html
    # Imagen_float2D = pylab.imread("20_popc_cho009-1.tif").astype(float) # Creates a 2D array
    #Reference http://nbviewer.ipython.org/github/profjsb/python-bootcamp/blob/master/Lectures/05_NumpyMatplotlib/Exercise-FourierImageDenoiseSolution.ipynb
    fig = plt.figure(1, figsize=(10.5,8.5))

    imshow(abs_array_filtrado[:,:], cmap = cm.Greys_r)

    plt.title('Imagen filtrada - Lowpass filter - Gausiano', fontsize=12)
    
```

```python
def funcion_filtro_fuerte(filename):

    shift_array = shift_fourier(filename)
    #shift_array = procesar_imagen(filename)
    r = shift_array.shape[0]
    c = shift_array.shape[1]
    
    #r,c = shape(shift_array)
    
    M = numpy.ones((r,c))
    
    centro_eje_x = c/2
    centro_eje_y = r/2
    
    limite_frecuencias = 10
    
    t = 2*limite_frecuencias
    
    for i in range(1,r):
        for j in range(1,c):
            radio_al_centro = numpy.sqrt( (centro_eje_x - j)**2 + (centro_eje_y - i)**2 )
            
            if (radio_al_centro < limite_frecuencias):
                M[i][j] = 1 - ( math.exp(-(radio_al_centro**2)/(t**2) ) )
    
    
    M = scipy.misc.toimage(M)
    
    array_filtrado = shift_array * M
    
    array_filtrado_inversa = fftim.ifft2(array_filtrado)
    
    abs_array_filtrado = abs(array_filtrado_inversa)
    
    imagen = scipy.misc.toimage(abs_array_filtrado)
    
    
    #Reference http://docs.opencv.org/modules/highgui/doc/reading_and_writing_images_and_video.html
    # Imagen_float2D = pylab.imread("20_popc_cho009-1.tif").astype(float) # Creates a 2D array
    #Reference http://nbviewer.ipython.org/github/profjsb/python-bootcamp/blob/master/Lectures/05_NumpyMatplotlib/Exercise-FourierImageDenoiseSolution.ipynb
    fig = plt.figure(1, figsize=(10.5,8.5))
    plt.title('Imagen filtrada _ Highpass filter -gaussiano', fontsize=20 )
    imshow(abs_array_filtrado[:,:], cmap = cm.Greys_r)
 
    
```

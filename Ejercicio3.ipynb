```python
f = open('circulos.dat', 'r') 
lines = f.readlines()
f.close()
```

```python
 x1 = []
y1 = []
a=[]
```

```python

for line in lines:
    p = line.split()
    x1.append(float(p[0]))
    y1.append(float(p[1]))
    a.append([float(p[0]),float(p[1])])
```

The file circulos.dat contains positions in the x-y plane. Those points correspond to the superposition of differente circles and points randomly distributed from an homogeneous distribution. This program finds the diameter of those circles using Fourier Transform.

We accomplish the goal using the autocorrelation function. It is the cross correlation of a signal with itself. In other words, it expresses the similarity between the function at different time. It allows us to find repeating patterns in a function. The mathematical definition is as follow:

$C(t)=\int^{-\infty}_{\infty} f(\tau)\bar{f}(\tau-t)d\tau$

We can use Wiener-Kinchin Theorem to find the correlation function since it tells us that the autocorrelation is simpy the Fourier transform of the absolute square of the Fourier Transform of the Function. Mathematically,

$C(t)=\mathscr{F}_{v}[|E_{v}|^2]$

The previous code finds the autocorrelation using the Wiener Khinchin theorem with normalization (Using the maximum value of the autocorrelation).

```python
#Function finds the maximas of a two dimensional array. It was adapted from http://stackoverflow.com/questions/3986345/how-to-find-the-local-minima-of-a-smooth-multidimensional-array-in-numpy-efficie
#Which finds the minima of a two dimensional array
def local_maxima(array2d):
    return ((array2d >= np.roll(array2d,  1, 0)) &
            (array2d >= np.roll(array2d, -1, 0)) &
            (array2d >= np.roll(array2d,  1, 1)) &
            (array2d >= np.roll(array2d, -1, 1)))
```

```python
filename = 'prueba.tif'
imagen = Image.open(filename).convert('L') #http://effbot.org/imagingbook/image.htm
    
array_imagen = numpy.asarray(imagen) # Convert the input to an array. - http://docs.scipy.org/doc/numpy/reference/generated/numpy.asarray.html

plt.imshow(array_imagen)
```

```python
f=fft2(array_imagen)
f=f*np.conj(f)
f = ifft2(f)

F_Transform=f
r,c = shape(F_Transform)

ff_order = F_Transform.copy()

size_filas = r # Gives the number of rows in the 2d Array
size_columnas = c# Gives the number of columns in the 2d Array

ff_order[ size_filas/2:size_filas , size_columnas/2:size_columnas ] = F_Transform[0:size_filas/2 , 0:size_columnas/2]
ff_order[ 0:size_filas/2 , 0:size_columnas/2] = F_Transform[size_filas/2:size_filas , size_columnas/2:size_columnas ]
ff_order[ 0:size_filas/2 , size_columnas/2:size_columnas ] = F_Transform[size_filas/2:size_filas , 0:size_columnas/2]
ff_order[size_filas/2:size_filas , 0:size_columnas/2] = F_Transform[0:size_filas/2 , size_columnas/2:size_columnas ]
f=ff_order

f = real(f)

f = f/amax(amax(f))
plt.imshow(f)
```

The previous plot corresponds to the autocorrelation function. We want to find the local maxima, for this we use a gaussian filter to smooth the image. The previous based on that we need to find the strongest intensities. In the autocorrelation function we can see that there are 7 points of strongest intensities. 

```python
H1 = gaussian_filter(f, 3, mode='constant')
detected_peaks = local_maxima(H1)
plt.imshow(detected_peaks)
print tuple(zip(*np.where(detected_peaks  == True))) #Give us the cells of the maximum point
```

The distance between nearest maxima corresponds to the diameter of the circles (withouth taking into account the global maxima). In this case, the diameter of each circle will be 16 pixels. Which agrees with the image!!!Now we know how to calculate diameter of circles.

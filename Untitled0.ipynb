```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fft, fftfreq, fft2
```

```python
f = open('circulos.dat', 'r') 
lines = f.readlines()
f.close()
```

```python
 x1 = []
y1 = []
a=[]
```

```python

for line in lines:
    p = line.split()
    x1.append(float(p[0]))
    y1.append(float(p[1]))
    a.append([float(p[0]),float(p[1])])
```

```python
plt.scatter(x1,y1,s=0.1)
```

```python
n_points=size(x1)
x1_fft = fft(x1) / n_points 
y1_fft = fft(x1) / n_points 
freq = fftfreq(n_points, 1) 
plt.plot(freq,np.abs(y1_fft))
```

```python
b=fft2(a)
```

```python
for i in range(size(x1)):
    x1[i]=b[i][0]
    y1[i]=b[i][1]
```

```python
plt.scatter(np.absolute(x1),np.absolute(y1),s=0.1)
```

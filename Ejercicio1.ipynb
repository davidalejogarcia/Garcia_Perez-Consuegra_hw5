## Here I order the Fourier Transform

```python
ff_order = F_Transform.copy()

size_filas = r # Gives the number of rows in the 2d Array
size_columnas = c# Gives the number of columns in the 2d Array

ff_order[ size_filas/2:size_filas , size_columnas/2:size_columnas ] = F_Transform[0:size_filas/2 , 0:size_columnas/2]
ff_order[ 0:size_filas/2 , 0:size_columnas/2] = F_Transform[size_filas/2:size_filas , size_columnas/2:size_columnas ]
ff_order[ 0:size_filas/2 , size_columnas/2:size_columnas ] = F_Transform[size_filas/2:size_filas , 0:size_columnas/2]
ff_order[size_filas/2:size_filas , 0:size_columnas/2] = F_Transform[0:size_filas/2 , size_columnas/2:size_columnas ]

fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.set_zlim(0, 0.1*(10**7))
p = ax.plot_surface(X, Y, abs(ff_order), rstride=4, cstride=4, linewidth=0)

plt.show()

print abs(ff_order)
```

```python
fraction = 5 # Limit value of frecuency
w = 1

size_filas = r # Gives the number of rows in the 2d Array
size_columnas = c# Gives the number of columns in the 2d Array
```

```python
#ff = F_Transform.copy() #makes a copy of the fourier transform array and calls it ff

# Calculo los centros (0,0) del array 2d de ff
centro_x = size_columnas/2
centro_y = size_filas/2

plt.axes()

circle = plt.Circle((centro_x, centro_y), radius=fraction, fc='y')
plt.gca().add_patch(circle)

#plt.axis()
pylab.ylim([0,size_filas])
pylab.xlim([0,size_columnas])
plt.axis('scaled')
plot(centro_x,centro_y,'ro')
plt.show()

pi = 3.14
ff = ff_order
for i in range (size_filas):
    for j in range (size_columnas):
        a = abs(centro_x - j)
        b = abs(centro_y - i)
        radio2 = ((a)**2 + (b)**2)**0.5
        
        if(radio2 < fraction - w):
            ff[i][j] = ff[i][j]*0
        if(radio2 > fraction + w):
            ff[i][j] = ff[i][j]*1
            
        else:
            ff[i][j] = ff[i][j]*( 0.5*( (1) - ( sin(pi*( (ff[i][j]) - fraction )/ (2*w) ) ) ) )
ff_order = ff
print ff_order
```

## F_order back. This part organizes the F_order array back to the original way so as to prepare it to trasnforma back into the real space

```python
ff_order2 = ff_order

ff_order2[ size_filas/2:size_filas , size_columnas/2:size_columnas ] = ff_order[0:size_filas/2 , 0:size_columnas/2]
ff_order2[ 0:size_filas/2 , 0:size_columnas/2] =  ff_order[size_filas/2:size_filas , size_columnas/2:size_columnas ]
ff_order2[ 0:size_filas/2 , size_columnas/2:size_columnas ] =  ff_order[size_filas/2:size_filas , 0:size_columnas/2]
ff_order2[size_filas/2:size_filas , 0:size_columnas/2] =  ff_order[0:size_filas/2 , size_columnas/2:size_columnas ]

print ff_order2
```
